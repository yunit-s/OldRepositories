## 연습문제
### Titanic - Machine Learning from Disaster

```python
# Module 불러오기
import numpy as np
import pandas as pd

# Raw Data Loading
train = pd.read_csv('./data/titanic/train.csv')
display(train.head())
# Survived
#     0 : 사망, 1 : 
# Pclass
#     객실 등급
# Name
#     이름
# Sex
#     성별
# SibSp
#     형제, 배우자 인원수
# Parch
#     부모, 자식 인원수
# Cabin
#     캐비넷 번호
# Embarked
#     승선지

# 종속변수 설정
#     생존여부. Survived
#     나머지는 모두 독립변수

# 독립변수 중 종속변수에 영향을 미치지 않는 것들 찾아서 제거하기
#     원래는 상관대수 같은 것들을 조사해야 한다. 실제로 수학 식이 있다.
#     PassengerId, Pclass, Name(의미를 더 찾을 수는 있으나 우리는 제외시키자), Ticket, Cabin
#     Fare같은 경우, Pclass와 비슷하게 가기 때문에 제외시키자.
train.drop(['PassengerId', 'Name', 'Ticket', 'Fare', 'Cabin'], axis=1, inplace=True)
#     우리는 지금 열을 지울 거야. '행 과 열'. '0 과 1'. 열이 1 이라고 외워도 된다.
display(train.head())

# 머신러닝에 글자는 넣을 수 없다.
#     0 : male, 1 : female
my_map = {'male' : 0, 'female': 1} # dictionary
train['Sex'] = train['Sex'].map(my_map)
# map()
#     원본 시리즈에서, () 안의 딕셔너리 데이터에 맞춰서 Series의 값을 바꿔준다.
display(train.head())

# 가족 데이터 처리
train['Family'] = train['SibSp'] + train['Parch']
train.drop(['SibSp', 'Parch'], axis=1, inplace=True)
display(train.head())

# 결측치 처리
display(train.info())
# Age
#     NaN이 꽤 있다.
# Embarked
#     NaN이 2개 있다.
# 이번엔 지우지 않고, NaN에 값을 채워보자.

# 결측치 처리 - Embarked
# 가장 빈도가 많은 걸 찾아서 대체하자.
train['Embarked'] = train['Embarked'].fillna('Q') # NaN 값을 대체하기
display(train.info())
my_embarked_map = { 'C': 0, 'S': 1, 'Q': 2 }
train['Embarked'] = train['Embarked'].map(my_embarked_map)
display(train.head())

# 결측치 처리 - Age
# 삭제하면 데이터가 너무 적어진다. 다른 값으로 대체하자.
# 대표값 지정
#     평균
train['Age'] = train['Age'].fillna(train['Age'].mean())
display(train.info())

# Age만 범위가 크다.
# 정규화를 해도 된다. 그런데 정규화를 하면, 나중에 역정규화를 해야한다.
# 구간값으로 사용하는 게 더 좋을 것 같다.
#

# 8살 이하만 추출
train.loc[train['Age'] < 8, 'Age'] = 0
train.loc[(train['Age'] >= 8) & (train['Age'] < 20), 'Age'] = 1
train.loc[(train['Age'] >= 20) & (train['Age'] < 65), 'Age'] = 2
train.loc[train['Age'] >= 65, 'Age'] = 3

display(train.head())
```

```python
# Training Data Set

x_data = train.drop('Survived', axis=1, inplace=False).values
print(x_data)
t_data = train['Survived'].values.reshape(-1,1)
```

```python
# Model 만들기

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import Adam

# Model
model = Sequential()

# Model에 Layer 추가
# 독립변수의 개수만큼 input layer의 동그라미가 있어야 한다.
# 지금 독립변수 5개니까 동그라미 5개 사용
model.add(Flatten(input_shape=(5,)))
model.add(Dense(1, activation='sigmoid'))

# Model 설정
model.compile(optimizer=Adam(learning_rate=1e-2), loss='binary_crossentropy',
             metrics=['accuracy'])

# Model 학습
model.fit(x_data, t_data, epochs=500,
          validation_split=0.2,
          verbose=1)
# 출력
#     Epoch 500/500
#     23/23 [==============================] - 0s 3ms/step - loss: 0.4566 - accuracy: 0.8034 - val_loss: 0.3935 - val_accuracy: 0.8156
#     <keras.callbacks.History at 0x24ffd50d280>
# val_accuracy 값이 가장 중요한데, 0.82 정도를 벗어나지 못한다.
#     데이터의 개수가 적어서 이 정도가 한계인 것
```
